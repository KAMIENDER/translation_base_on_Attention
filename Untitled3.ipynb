```python
import torch
import torch.nn as nn
import numpy as np
import os
# 有gpu的情况下使用gpu
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
#device = torch.device('cpu')


class voc_dict():
    def __init__(self):
        self.word2idx = {}
        self.idx2word = {}
        self.idx = 0

    def add_words(self,word):
        if word not in self.word2idx:
            self.word2idx[word] = self.idx
            self.idx2word[self.idx] = word
            self.idx += 1

class build_dataset():
    def __init__(self):
        self.dictionary = voc_dict()
        self.dictionary.add_words('bos')
        self.dictionary.add_words('eos')
    
    def size(self):
        return len(self.dictionary.word2idx)

    def get_data(self,file,time_steps):
        dataset = []
        self.dictionary.add_words('padding')
        with open(file,'r',encoding='utf-8') as f:
            content = f.readlines()
            for line in content[:20]:
                words = line.strip().split()
                if len(words)>time_steps-2:
                    words = words[0:time_steps-2]
                words = ['bos']+words+['eos']
                idx = []
                for word in words:
                    self.dictionary.add_words(word)
                while (len(words)<time_steps):
                    words= words + ['padding']
                for word in words[0:time_steps]:
                    idx.append(self.dictionary.word2idx[word])
                dataset.append(idx)
        dataset = np.array(dataset)
#         print(dataset)
        # ids = torch.LongTensor(dataset)
        # print(ids.size())
        return dataset

    def seq2words(self,text):
        result = ''
        for i in text:
#             i = i.numpy()
            for j in i:
                if j!=2:
                    result+=self.dictionary.idx2word[j]+' '
            result+='\n'
        return result

zh_dict = build_dataset()
data_source = torch.LongTensor(zh_dict.get_data('source.txt',20))   
# print(data_source)
print(zh_dict.size())

en_dict = build_dataset()
data_target = torch.LongTensor(en_dict.get_data('target.txt',20))
print(data_target.size())
print(en_dict.size())

print(en_dict.seq2words(data_target[:2,:]))
print(zh_dict.seq2words(data_source[:2,:]))

input_size = zh_dict.size()
output_size = en_dict.size()
embed_size = 20
hidden_size = 32
num_layers = 1
num_epochs = 10
batch_size = 1
time_steps = 20
lr = 0.002
```

```python
class rnn(nn.Module):
    def __init__(self,input_size,embed_size, hidden_size, num_layers,output_size):
        super(rnn,self).__init__()
        self.embed = nn.Embedding(input_size, embed_size)
        self.lstm = nn.LSTM(embed_size, hidden_size,batch_first=True)
        self.linear = nn.Linear(hidden_size, output_size)

    def forward(self, x, h=None):
        x = self.embed(x)
        out,h_c = self.lstm(x,h)
#         print(h_c.size())
        out = out.reshape(out.size(0)*out.size(1),out.size(2))
        out = self.linear(out)
        return out,h_c


encoding = rnn(input_size, embed_size, hidden_size, num_layers,output_size).to(device)
print(encoding)
decoding = rnn(output_size, embed_size, hidden_size, num_layers,output_size).to(device)
print(decoding)
criterion = nn.CrossEntropyLoss()
encoder_optimizer = torch.optim.Adam(encoding.parameters(),lr=0.006)
decoder_optimizer = torch.optim.Adam(decoding.parameters(),lr=0.006)
# input = data_source[0:2,:].to(device)
# print(input.size())
# target = data_target[0:2,:].to(device)
# hidden = None
# x = input[:,[1]]
# print(x.size())
# output,h_m = encoding(x,hidden)
# hidden = h_m
# x = torch.LongTensor([[0],[0]])
# for i in range(target.size()[1]):
#     output,h_c = decoding(x,hidden)
#     x = torch.max(output, 1)[1].reshape(-1,1)
#     hidden = h_c

print(data_source.size())
for epoch in range(500): 
    for i in range(0,data_source.size(0)//batch_size-1): 
        input = data_source[i*batch_size:(i+1)*batch_size,:].to(device) 
        target = data_target[i*batch_size:(i+1)*batch_size,:].to(device) 
        hidden = None 
        output,h_m = encoding(input,hidden) 
        hidden = h_m
#         print(target.size())
        for j in range(target.size()[0]):
            x = target[[j],[0]].reshape(1,1)
#             print(x.size())
            result,h_c = decoding(x,hidden)
            for i in range(1,target.size()[1]-1):
                output,h_c = decoding(x,h_c)
    #             print(h_c.size())
                result=torch.cat((result,output),0)  
#         print(result.size())
#         print(target.size())
        loss = criterion(result,target[:,1:].reshape(-1) ) 
#         print(result.size())

        encoder_optimizer.zero_grad()
        decoder_optimizer.zero_grad()
        loss.backward()
        encoder_optimizer.step()
        decoder_optimizer.step()
#         print('Epoch: ', epoch, '| train loss: %.4f' % loss.data.cpu().numpy())
    if epoch%5 == 0:
        target = data_target
        input = data_source
        hidden = None 
        output,h_m = encoding(input,hidden) 
        hidden = h_m
#         print(target.size())
        for j in range(target.size()[0]):
            x = target[[j],[0]].reshape(1,1)
#             print(x.size())
            h_c = (hidden[0][:,[j],:],hidden[1][:,[j],:])
#             print(h_c[0].size())
            result,h_c = decoding(x,h_c)
            for i in range(1,target.size()[1]-1):
                output,h_c = decoding(x,h_c)
    #             print(h_c.size())
                result=torch.cat((result,output),0)  
            if j==0:
                out = result
            else:
                out = torch.cat((out,result),0)  
        print(out.size())
#         print(target.size())
        loss = criterion(out,target[:,1:].reshape(-1) ) 
        pred_y = torch.max(out, 1)[1].data.cpu().numpy().reshape(-1,19)
        target = target[:,1:].data.cpu().numpy().reshape(-1)
#         print(pred_y.shape)
#         print(target.shape)
        if epoch%20 == 0:
            ft = open('result/'+str(epoch)+'.txt','w',encoding='utf-8')
            ft.write(en_dict.seq2words(pred_y))
            ft.close()
        pred_y = pred_y.reshape(-1)

        accuracy = float((pred_y == target).astype(int).sum()) / float(target.size)
        print('Epoch: ', epoch, '| train loss: %.4f' % loss.data.cpu().numpy(), '| test accuracy: %.2f' % accuracy)
```
